```python
import numpy as np  
import matplotlib as plt
import pandas as pd 
from tensorflow.keras.models import Sequential, load_model
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import Normalization
import tensorflow as tf


pathname = 'admission_data.csv' # import the csv 
df = pd.read_csv(pathname) # reading it into a data frame 


normalized_data = df.copy().sample(frac=1) # making copy

values_to_normalize = ['GRE Score','TOEFL Score','CGPA'] # these are the features that I d

for x in values_to_normalize:             # a for loop to normalize some featurues with espcially large outputs 
    curr_min = normalized_data[x].min()
    curr_max = normalized_data[x].max()
    curr_range = normalized_data[x].max()-normalized_data[x].min()

    normalized_data[x]= (normalized_data[x]-curr_min)/curr_range
    

# these are the features that I decided to train
features_to_train = ['GRE Score','TOEFL Score','University Rating','SOP','LOR ','CGPA','Research']


model_inputs = normalized_data[features_to_train]

#this is the feature that I decided to predict 
model_outputs = normalized_data['Chance of Admit ']

#seperating out my data 
training_x = model_inputs[:300]
training_y = model_outputs[:300]

cross_valid_x = model_inputs[301:400]
cross_valid_y = model_outputs[301:400]

test_set_x = model_inputs[401:]
test_set_y = model_outputs[401:]


```

```python
 # generating the model with 20 neurons in the first hidden layer, 15 in the second, 6 in the third and 1 in the output layer 
model = Sequential([
    Dense(units=20, input_shape=[7], activation = 'relu'),
    Dense(units=15, activation = 'relu'),
    Dense(units=6, activation = 'relu'),
    Dense(units=1, activation = 'sigmoid')
])
```

```python
# using mean square error for our loss function because an error rate of 5% is acceptable for this case 
# using adams algorithm to help the algorthim converge a bit faster
# using mean absolute error to get a generic idea of where the model is at 


model.compile(loss='mse',optimizer='adam',metrics=[tf.keras.metrics.MeanAbsoluteError()]) 
```

```python
history = model.fit(training_x,training_y,epochs=500,batch_size=32)
```

```python
y_hat = model.predict(test_set_x).transpose()

difference_test= (test_set_y - y_hat[0]).abs()

y_valid = model.predict(cross_valid_x).transpose()

difference_valid= (cross_valid_y - y_valid[0]).abs()

#examining the absolute squared error of our model on data it has not seen before usually about 4-5% off

print(difference_test.mean())
print(difference_valid.mean())
```
